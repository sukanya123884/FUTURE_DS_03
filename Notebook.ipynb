```python
# Create a column 
def generate_comment(rating):
    if rating >= 8:
        return "The course was excellent and  relevant to my career."
    elif rating >= 5:
        return "The course was okay but could be improved."
    else:
        return "The Course was very bad and poor."
df['Feedback'] = df['Course recommendation based on relevance'].apply(generate_comment)
```

```python
# Define a function for sentiment
def get_sentiment(text):
    score = TextBlob(text).sentiment.polarity
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'
df['Sentiment'] = df['Feedback'].apply(get_sentiment)
print(df[['Feedback', 'Sentiment']].head(10))
```
